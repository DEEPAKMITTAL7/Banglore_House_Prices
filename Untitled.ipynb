We can see that there is a range values also, lets take the mean and fix this feature as well.

```python
def convert_sqft(x):
    test = x.split('-')
    if len(test)==2:
        return ( float(test[0]) + float(test[1]) ) / 2
    try:
        return float(x)
    except:
        return None
```

```python
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft)  #some values had 34.3sqft has been maeked as None
```
